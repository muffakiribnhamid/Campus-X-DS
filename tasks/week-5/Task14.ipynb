```python
# Find the nearest element in the array to a given integer
import numpy as np
arr = np.array([10, 17, 24, 31, 38, 45, 52, 59])
given_element = 30

diff = np.abs(arr - given_element)

print(diff)
 
idx = np.argmin(diff)

print(idx)

nearest = arr[idx]

print(nearest)
```

```python
# Q-3Write a function to create a 2d array with 1 on the border and 0 inside. 
# Take 2-D array shape as (a,b) as parameter to function.



def create_arr(a,b):
    arr = np.zeros((a,b),dtype=int)
    arr[0,:] = 1
    arr[-1,:] = 1
    arr[:,0] = 1
    arr[:,-1] = 1
    
    return arr

create_arr(4,4)
    
```

```python
 # Consider a random integer (in range 1 to 100) vector with shape (10,2) 
# representing coordinates, and coordinates of a point as array is given.
# Create an array of distance of each point in the random vectros from the given point. Distance array should be interger type


arr = np.random.randint(1,101,size=(10,2))
ref_point = np.array([2,3]) 
distances = np.sqrt(np.sum((arr - ref_point) ** 2, axis=1))

distances = distances.astype(int)

print("Points: ",arr)
print("Distances: ",distances)

```

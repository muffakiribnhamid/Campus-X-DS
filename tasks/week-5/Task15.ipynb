```python
# Create a random 3x4 matrix with value between 0-100. And perform below tasks
# i. Sort this matrix. np.sort()
# ii. Sort this matrix based on values in 2nd column.
# iii. Sort this matrix based on max value in each row.
# iv. Sort based on elements value.
```

```python
arr = np.random.randint(1,100,size=(3,4))

print(arr)

print("=========Sorted=======")


sorted = np.sort(arr)

print(sorted)

print("======Sorted By Second Col=======")


#sort by second columns

sorted_column = arr[arr[:,1].argsort()]

print(sorted_column)
print("========Sort this matrix by max val in each row ============")
sorted_row = arr[np.max(arr,axis=1).argsort()]
print(sorted_row)


sorted_elementsval = np.sort(arr, axis=None).reshape(arr.shape)
print(sorted_elementsval)

```

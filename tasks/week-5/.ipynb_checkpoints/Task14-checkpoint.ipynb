```python
# Find the nearest element in the array to a given integer
import numpy as np
arr = np.array([10, 17, 24, 31, 38, 45, 52, 59])
given_element = 30

diff = np.abs(arr - given_element)

print(diff)
 
idx = np.argmin(diff)

print(idx)

nearest = arr[idx]

print(nearest)
```

```python
# Q-4: Your are given an array
# which is havig some nan value. You job is to fill those nan values with most common element in the array.


arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

clean = arr[~np.isnan(arr)]

values,counts = np.unique(clean,return_counts=True)

most_common = values[counts.argmax()]

arr[np.isnan(arr)] = most_common

print(arr)
```

```python
# Q-9: Write a program implement Boradcasting Rule to check if two array can be added or not.
# Given tuples of shapes.

# shape of a- (3,2,2)
# shape of b- (2,2)

# check_broadcast(a, b) -> return Boolean (True if can broadcasted, False other wise.)



def check_broadcast(a,b):
    len_diff = len(a) - len(b)

    if len_diff > 0:
        a = (1,) * len_diff + b
    elif len_diff < 0:
        b = (1,) * (-len_diff) + a

    #check for each dimension
    for da,db in zip(a,b):
        if da != db and da!= 1 and db != 1:
            return False
    return True


a = (3,2,2)
b = (2,2)

check_broadcast(a,b)
```

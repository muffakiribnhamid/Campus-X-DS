```python
"""

##`Q-1:` Rectangle Class
1. Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.

2. Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of  rectangle.

3. Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.

"""




class Rectangle:

    """
    Rectangle class which allows to build a rectangle with length and width attributes
    """

    def __init__(self,length,width):
        self.length = length
        self.width = width
        self.Display()


    def Perimeter(self):
        perimeter = (self.length + self.width) * 2
        return perimeter

    def Area(self):
        area = self.length * self.width
        return area

    def Display(self):
        print("Length: ", self.length)
        print("Width: ", self.width)
        print("Area: ", self.Area())
        print("Perimeter: ", self.Perimeter())



rect1 = Rectangle(10,20)
```

```python
#Question 2 
class BankAccount:
    def __init__(self, accountNumber, name, balance):
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} deposited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdrawal(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"{amount} withdrawn successfully.")
        else:
            print("Invalid or insufficient funds.")

    def bank_fees(self):
        fee = self.balance * 0.05
        self.balance -= fee
        print(f"Bank fee of {fee} applied.")
        return self.balance

    def display(self):
        print("Account Number:", self.accountNumber)
        print("Account Holder:", self.name)
        print("Balance:", self.balance)


# Example usage:
bank1 = BankAccount(1234, "X", 200)
bank1.deposit(100)
bank1.withdrawal(50)
bank1.bank_fees()
bank1.display()
```

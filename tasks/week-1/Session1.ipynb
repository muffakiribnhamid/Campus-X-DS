```python
# Q6:- Write a program that will tell the number of dogs and chicken 
# are there when the user will provide the value of total heads and legs.


heads = int(input("Enter number of heads: "))
legs = int(input("Enter number of legs:"))

dogs = (legs - (2 * heads)) // 2
chickens = heads - dogs


if dogs >= 0 and chickens >= 0 and (legs - 2 * heads) % 2 == 0:
    print("Chicks : ", chickens)
    print("Dogs : ", dogs)
else:
    print("Invalid Input")
    
    
```

```python
# Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

n = int(input("Enter first n: "))
sum = 0 

for i in range(1,n+1):
    sum = sum + i ** 2

print(sum)
```

```python
 # Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

a1 = int(input("Enter first term: "))
a2 = int(input("Enter second term: "))
n = int(input("Enter term number (n): "))

d = a2 - a1
an = a1 + (n - 1) * d

print("Nth term is:", an)
```

```python
### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

n1 = int(input("Enter first numerator : "))
n2 = int(input("Enter second numerator : "))
d1 = int(input("Enter first denominator : "))
d2 = int(input("Enter second denominator : "))

num = (n1 * d2) + (n2 * d1) 
den = d1 * d2

print("Sum", num, "/", den)

```

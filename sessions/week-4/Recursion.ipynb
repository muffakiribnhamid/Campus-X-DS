```python
#using loops
def multiply(a,b):

    result = 0
    for i in range(b):
        result = result + a
    return result

multiply(4,3)
```

```python
#using recursion

def mul(a,b):
    if b == 1:
        return a
    else:
        return a + mul(a,b-1)


mul(7,8)
```

```python
#factorial using recursion

def fact(num):
    if num == 1:
        return num
    else:
        return num * fact(num - 1)


fact(5)
```

```python
#palindrome

def pal(str):
    if len(str) <= 1:
        print("It's a Palindrome")
    else:
        if str[0] == str[0]:
            pal(str[1:-1])
        else:
            print("Not a palindrome")


pal("Salman")
```

```python
#The rabbit problem | fibonacci series

def fib(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)


print(fib(12))
```

```python
#memoization

def memo(m,d):
    if m in d:
        return d[m]
    else:
        d[m] = memo(m-1,d) + memo(m-2,d)
        return d[m]

d = {0:1,1:1}

print(memo(500,d))
```

```python
#write lines
L = ["Hello","World","I","Miss","Her"]

g = open("sample.txt","a")
g.writelines(L)
g.close()
```

```python
#reading 
f = open("sample.txt","r")
s = f.read()
print(s)
f.close()
```

```python
#reading upto n chars
f = open("sample.txt","r")
s = f.read(10)
print(s)
f.close()
```

```python
#readline()
f = open("sample.txt","r")
print(f.readline(),end=' ')
print(f.readline(),end=' ')
f.close()
```

```python
f = open("sample.txt","r")
while f.readline() != '':
    print(f.readline())

f.close()
```

```python
f = open("sample.txt","w")
f.writelines("Muba")
print(f.tell())
f.close()
```

```python
#with -> context manager
#it closes the file as soon as usage is over

#with
with open("muffakir.txt","w") as f:
    f.write("Muba")
```

```python
big_l = ["Hello World" for i in range(1,100)]

with open("test.txt","w") as f:
    f.writelines(big_l)


```

```python
with open("test.txt","r") as f:
    chunk_size = 10
    
    while len(f.read(chunk_size)) > 0:
        print(f.read(chunk_size),end="***")
        f.read(chunk_size)
```

```python
with open("test.txt","r") as f:
    print(f.read(10))
    print(f.tell())
    f.seek(15)
    print(f.read(10))
    print(f.tell())
    
    
    
```

```python
class Person:

  def __init__(self,fname,lname,age,gender):
    self.fname = fname
    self.lname = lname
    self.age = age
    self.gender = gender

# format to printed in
# -> Nitish Singh age -> 33 gender -> male
```

```python
person = Person("Muffakir","Hamid",18,"Male")
```

```python
import json

def show_object(person):
    if isinstance(person,Person):
        return f"{person.fname} {person.lname} age -> {person.age} gender -> {person.gender}"

with open("demo.json", "w") as f:
    json.dump(person,f,default=show_object)
```

```python
#Pickling
#Pickling is the process whereby a Python object hierarchy is converted into a byte stream, and `unpickling`
#is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back
#into an object hierarchy.
```

```python
class Person:

  def __init__(self,name,age):
    self.name = name
    self.age = age

  def display_info(self):
    print('Hi my name is',self.name,'and I am ',self.age,'years old')
```

```python
p = Person("Muffakir",19)
```

```python
#pickle dump
import pickle

with open("person.pkl", "wb") as f:
    pickle.dump(p,f)
```

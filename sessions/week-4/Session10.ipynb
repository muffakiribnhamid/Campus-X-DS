```python
# File Handling + Serialization & Deserialization

"""
types of data used for I/O
- Text : "12345"
- Binary : 12345

Hence there are two types of files
- Text File
-Binary File


How file I/O is done in prog lang's

- Open a file
- Read/Write a file
- Close the file
"""

```

```python
#case 1 - if file is not present

f = open("sample.txt","w")
f.write("Hello World")
f.write("\nHow are you?")
f.close()
#after closing the file we cant' write to it.
```

```python
#case 2 - if file is already present
#same process. it replaces the content.
```

```python
#introducing append mode

f = open("sample.txt","a")
f.write("\nI am a geek\n")
f.close()
         
```

```python
#write lines
L = ["Hello","World","I","Miss","Her"]

g = open("sample.txt","a")
g.writelines(L)
g.close()
```

```python
#reading 
f = open("sample.txt","r")
s = f.read()
print(s)
f.close()
```

```python
#reading upto n chars
f = open("sample.txt","r")
s = f.read(10)
print(s)
f.close()
```

```python
#readline()
f = open("sample.txt","r")
print(f.readline(),end=' ')
print(f.readline(),end=' ')
f.close()
```

```python
f = open("sample.txt","r")
while f.readline() != '':
    print(f.readline())

f.close()
```

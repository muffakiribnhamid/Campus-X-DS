```python
#  What is a Generator
# Python generators are a simple way of creating iterators. 
```

```python
def gen_demo():
    
    yield "first statement"
    yield "second statement"
    yield "third statement"
    
gen = gen_demo()

for i in gen:
    print(i)
```

```python
def square(num):
    for i in range(1,num+1):
        yield i**2
gen = square(10)

print(next(gen))
print(next(gen))
print(next(gen))

for i in gen:
    print(i)
```

```python
#range function using generator

def mera_range(start,end):
    for i in range(start,end+1):
        yield i
```

```python
for i in mera_range(1,100):
    print(i)
```

```python
# Pandas Series
# A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.
```

```python
#importing pandas
import numpy as np
import pandas as pd
```

```python
#series from list
```

```python
#strings
country = ["India","Pak","USA","Nepal"]
pd.Series(country)
```

```python
#intergers
runs = [12,23,243,2,232,101]
pd.Series(runs)
```

```python
#customs index
marks = [100,90,67,100,100,56]
subjects = ["Science","English","Urdu","Math","CS","Hindi"]

pd.Series(marks,index=subjects)
```

```python
#setting a name
marks = [100,90,67,100,100,56]
subjects = ["Science","English","Urdu","Math","CS","Hindi"]

pd.Series(marks,index=subjects,name="Muffakir ke marks")
```

```python
#series from dict

marks = {
    "maths":67,
    "english":100,
    "science":98,
    "urdu":100,
    "sst":100
}

marks_series = pd.Series(marks,name="X ke marks")
marks_series
```

```python
#SERIES Attributes:
```

```python
#1.size -> returns size of series or returns items
marks_series.size
```

```python
#2-dtype
marks_series.dtype
```

```python
#is_unique: tells if all items in series are unique

marks_series.is_unique
```

```python
#index -> gives back index array of all indexes

marks_series.index
```

```python
#values -> gives back array of values
marks_series.values
```
